## PCA 4d

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Select the features for evaluation
features = [
    'danceability', 'energy', 'key', 'loudness', 'mode',
    'speechiness', 'acousticness', 'instrumentalness',
    'liveness', 'valence', 'tempo', 'duration_ms'
]
data = df[features]

# Standardize the features
scaler = StandardScaler()
data_std = scaler.fit_transform(data)

# Perform PCA to reduce the dimensions to 4
pca = PCA(n_components=4)
data_pca = pca.fit_transform(data_std)

# Find the optimal number of clusters using the silhouette algorithm
silhouette_scores = []
max_clusters = 10

for n_clusters in range(2, max_clusters + 1):
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
    labels = kmeans.fit_predict(data_pca)
    silhouette_scores.append(silhouette_score(data_pca, labels))

optimal_clusters = np.argmax(silhouette_scores) + 2  # +2 since the range starts from 2
print(f"The optimal number of clusters is: {optimal_clusters}")

# Perform K-means++ clustering on the PCA-transformed data
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', random_state=42)
labels = kmeans.fit_predict(data_pca)

# Create a new DataFrame with PCA components and labels
data_pca_df = pd.DataFrame(data_pca, columns=['PC1', 'PC2', 'PC3', 'PC4'])
data_pca_df['Cluster'] = labels

# Use a pairplot to visualize the relationships between pairs of components
sns.pairplot(data_pca_df, hue='Cluster', diag_kind='hist', palette='bright')
plt.show()
```

## experiment with 3 reduction algorithms: 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import MDS, TSNE
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Select the features for evaluation
features = [
    'danceability', 'energy', 'key', 'loudness', 'mode',
    'speechiness', 'acousticness', 'instrumentalness',
    'liveness', 'valence', 'tempo', 'duration_ms'
]
data = df[features]

# Standardize the features
scaler = StandardScaler()
data_std = scaler.fit_transform(data)

# Apply PCA, MDS, and t-SNE dimensionality reduction algorithms
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_std)

mds = MDS(n_components=2, random_state=42)
data_mds = mds.fit_transform(data_std)

tsne = TSNE(n_components=2, random_state=42)
data_tsne = tsne.fit_transform(data_std)

# Find the optimal number of clusters for each reduced dataset
reduced_data = {'PCA': data_pca, 'MDS': data_mds, 't-SNE': data_tsne}
optimal_clusters = {}

for method, data_reduced in reduced_data.items():
    silhouette_scores = []
    max_clusters = 10

    for n_clusters in range(2, max_clusters + 1):
        kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
        labels = kmeans.fit_predict(data_reduced)
        silhouette_scores.append(silhouette_score(data_reduced, labels))

    optimal_clusters[method] = np.argmax(silhouette_scores) + 2

print("Optimal number of clusters for each method:", optimal_clusters)

# Perform K-means++ clustering and plot the results for each dimensionality reduction algorithm
for method, data_reduced in reduced_data.items():
    kmeans = KMeans(n_clusters=optimal_clusters[method], init='k-means++', random_state=42)
    labels = kmeans.fit_predict(data_reduced)

    data_reduced_df = pd.DataFrame(data_reduced, columns=['Component 1', 'Component 2'])
    data_reduced_df['Cluster'] = labels

    plt.figure(figsize=(10, 7))
    sns.scatterplot(x='Component 1', y='Component 2', hue='Cluster', data=data_reduced_df, palette='bright', legend='full', s=100)
    plt.title(f'{method} 2D Plot with K-means Clustering')
    plt.show()
```

## try T-sne reduction 3 components and interactive plot

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Select the features for evaluation
features = [
    'danceability', 'energy', 'key', 'loudness', 'mode',
    'speechiness', 'acousticness', 'instrumentalness',
    'liveness', 'valence', 'tempo', 'duration_ms'
]
data = df[features]

# Standardize the features
scaler = StandardScaler()
data_std = scaler.fit_transform(data)

# Apply t-SNE with 3 components
tsne = TSNE(n_components=3, random_state=42)
data_tsne = tsne.fit_transform(data_std)

# Find the optimal number of clusters using the silhouette algorithm
silhouette_scores = []
max_clusters = 10

for n_clusters in range(2, max_clusters + 1):
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
    labels = kmeans.fit_predict(data_tsne)
    silhouette_scores.append(silhouette_score(data_tsne, labels))

optimal_clusters = np.argmax(silhouette_scores) + 2
print(f"The optimal number of clusters is: {optimal_clusters}")

# Perform K-means++ clustering on the reduced data
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', random_state=42)
labels = kmeans.fit_predict(data_tsne)

# Create an interactive 3D plot using Plotly
fig = go.Figure()

for cluster in range(optimal_clusters):
    cluster_indices = np.where(labels == cluster)
    cluster_data = data_tsne[cluster_indices]

    fig.add_trace(go.Scatter3d(
        x=cluster_data[:, 0],
        y=cluster_data[:, 1],
        z=cluster_data[:, 2],
        mode='markers',
        marker=dict(size=4, opacity=0.8),
        name=f'Cluster {cluster}'
    ))

fig.update_layout(
    scene=dict(
        xaxis_title='t-SNE Component 1',
        yaxis_title='t-SNE Component 2',
        zaxis_title='t-SNE Component 3'
    ),
    title='3D t-SNE Plot with K-means Clustering',
    showlegend=True
)

fig.show()
```
